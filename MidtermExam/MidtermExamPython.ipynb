# ChatGpt Code:

    for key in mapping:
        text = text.replace(key, mapping[key])
    return text

# Critique:
Correctness: Works for small inputs but repeatedly recreates strings for each key.

• Complexity: O(n * m) where m = number of mapping entries — inefficient.

• Robustness: No input validation for types; assumes all lowercase.

```python
#Improved Code
```

```python

def translate(text, mapping):
    """Translate text into leetspeak using a given character mapping."""
    if not isinstance(text, str) or not isinstance(mapping, dict):
        return "Invalid input"
    
    result = []
    for ch in text.lower():
        result.append(mapping.get(ch, ch))  # use .get() for safe lookup
    return "".join(result)

# Test cases
mapping = {'a':'4','e':'3','i':'1','o':'0','s':'5','t':'7'}
print(translate("I like to eat eggs and spam.", mapping))
print(translate("HELLO WORLD", mapping))
print(translate(12345, mapping))  # Invalid input
```

```python

def check_password(pw):
    # Track if password meets each rule
    has_upper = False
    has_lower = False
    has_digit = False
    has_symbol = False
    has_no_spaces = True

    for ch in pw:
        if ch.isupper():
            has_upper = True
        elif ch.islower():
            has_lower = True
        elif ch.isdigit():
            has_digit = True
        elif ch in "!@#$%^&*":
            has_symbol = True
        elif ch == " ":
            has_no_spaces = False

    long_enough = len(pw) >= 10

    # Check if all rules are satisfied
    is_valid = all([has_upper, has_lower, has_digit, has_symbol, has_no_spaces, long_enough])

    return {
        "Length (>=10)": long_enough,
        "Has Uppercase": has_upper,
        "Has Lowercase": has_lower,
        "Has Digit": has_digit,
        "Has Symbol": has_symbol,
        "No Spaces": has_no_spaces,
        "Valid Password": is_valid
    }


# --- Main Program Loop ---
while True:
    pw = input("Enter your password (Q to quit): ")

    if pw.lower() == "q":
        print("You chose to quit. Program ended.")
        break

    result = check_password(pw)

    print("\nPassword Check Results:")
    for rule, passed in result.items():
        print(rule + ":", passed)

    if result["Valid Password"]:
        print("✅ Password accepted! Good job!")
        break
    else:
        print("❌ Password not accepted. Try again.\n")
```

I didn’t use random seeding since it wasn’t part of the lessons, so the result changes slightly each run — showing real randomness.

```python
# Improved Code
```

```python
import random

def roll():
    """Simulate rolling one fair die (1–6)."""
    return random.randint(1, 6)

def dice_simulation(trials=100000):
    """Simulate rolling two dice and estimate the probability that their sum equals 7."""
    success = 0

    for _ in range(trials):
        die1 = roll()
        die2 = roll()

        if die1 + die2 == 7:
            success += 1

    # Calculate estimated probability
    return success / trials


# --- Main Program ---
estimated = dice_simulation()
exact = 6 / 36  # There are 6 ways to get 7 out of 36 possible outcomes

print(f"Estimated Probability of sum=7: {estimated:.5f}")
print(f"Exact Probability: {exact:.5f}")
```

```python

class Vehicle:
    def __init__(self, color, mileage, fuel_liters):
        self.color = color
        self.mileage = mileage
        self.fuel_liters = fuel_liters

    # Drives the vehicle and updates mileage and fuel
    def drive(self, km, km_per_liter):
        fuel_needed = km / km_per_liter
        if fuel_needed > self.fuel_liters:
            print("Not enough fuel to drive that far.")
            km = self.fuel_liters * km_per_liter
            self.fuel_liters = 0
        else:
            self.fuel_liters -= fuel_needed
        self.mileage += km

    # Adds fuel to the tank
    def refuel(self, liters):
        if liters > 0:
            self.fuel_liters += liters
            print(f"Refueled {liters} liters. Total fuel: {self.fuel_liters}")
        else:
            print("Invalid amount of fuel.")

class Car(Vehicle):
    pass

class Truck(Vehicle):
    # Truck consumes more fuel (less efficient)
    def drive(self, km, km_per_liter):
        km_per_liter *= 0.8
        print("Truck drives with lower efficiency.")
        super().drive(km, km_per_liter)

# Test Script
car = Car("White", 15000, 50)
truck = Truck("Black", 30000, 70)

car.drive(120, 12)
truck.drive(120, 12)
car.refuel(10)
truck.refuel(20)

print(f"Car: Mileage={car.mileage}, Fuel={car.fuel_liters}")
print(f"Truck: Mileage={truck.mileage}, Fuel={truck.fuel_liters}")
```

```python
# Improved Code
```

```python
class Dog:
    def __init__(self, name, age, breed):
        self.name = name
        self.age = age
        self.breed = breed

    # Dog behavior
    def bark(self):
        print(f"{self.name} says Woof!")

    def __repr__(self):
        return f"{self.name} ({self.breed}, {self.age} yrs)"


class Kennel:
    def __init__(self):
        self.dogs = []

    # Add a dog to the kennel
    def add_dog(self, dog):
        if any(d.name == dog.name for d in self.dogs):
            print("A dog with this name already exists.")
        else:
            self.dogs.append(dog)
            print(f"{dog.name} added to the kennel.")

    # Remove a dog by name
    def remove_dog(self, name):
        for d in self.dogs:
            if d.name == name:
                self.dogs.remove(d)
                print(f"{name} removed from the kennel.")
                return
        print("Dog not found.")

    # Find the oldest dog
    def oldest_dog(self):
        if not self.dogs:
            print("No dogs in the kennel.")
            return None
        return max(self.dogs, key=lambda d: d.age)

    # Find all dogs of a given breed
    def find_by_breed(self, breed):
        found = [d for d in self.dogs if d.breed.lower() == breed.lower()]
        return found


# Test Script
dog1 = Dog("Buddy", 3, "Beagle")
dog2 = Dog("Max", 5, "Labrador")
dog3 = Dog("Luna", 2, "Beagle")

kennel = Kennel()
kennel.add_dog(dog1)
kennel.add_dog(dog2)
kennel.add_dog(dog3)

dog2.bark()
print("Oldest Dog:", kennel.oldest_dog())
print("Beagles:", kennel.find_by_breed("Beagle"))
kennel.remove_dog("Buddy")
```

```python
# Improved Code
```

```python
def cel_to_far(c: float) -> float:
    """
    Convert Celsius to Fahrenheit.
    
    Args:
        c (float): Temperature in Celsius.
    Returns:
        float: Temperature in Fahrenheit.
    """
    return (c * 9/5) + 32


def far_to_cel(f: float) -> float:
    """
    Convert Fahrenheit to Celsius.
    
    Args:
        f (float): Temperature in Fahrenheit.
    Returns:
        float: Temperature in Celsius.
    """
    return (f - 32) * 5/9


def main():
    try:
        print("Temperature Converter")
        choice = input("Convert from (C)elsius or (F)ahrenheit? ").strip().lower()

        if choice == "c":
            c = float(input("Enter temperature in Celsius: "))
            print(f"{c}°C = {cel_to_far(c):.2f}°F")
        elif choice == "f":
            f = float(input("Enter temperature in Fahrenheit: "))
            print(f"{f}°F = {far_to_cel(f):.2f}°C")
        else:
            print("Invalid choice. Please enter C or F.")
    except ValueError:
        print("Invalid input! Please enter numeric values only.")


# --- Simple Tests ---
assert cel_to_far(0) == 32
assert round(far_to_cel(32), 2) == 0
assert cel_to_far(-40) == far_to_cel(-40)  # -40 is same in both scales

if __name__ == "__main__":
    main()
```
